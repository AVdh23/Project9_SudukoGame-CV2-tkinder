## Grid Generator 

```python
import random

def generate_grid():
    grid = [[0] * 9 for _ in range(9)]
    fill_diagonal(grid)
    solve_grid(grid)
    remove_cells(grid)
    return grid

def fill_diagonal(grid):
    nums = list(range(1, 10))
    random.shuffle(nums)
    for i in range(0, 9, 3):
        fill_box(grid, i, i, nums)

def fill_box(grid, start_row, start_col, nums):
    num_index = 0
    for i in range(3):
        for j in range(3):
            grid[start_row + i][start_col + j] = nums[num_index]
            num_index = (num_index + 1) % 9

def solve_grid(grid):
    empty_cells = get_empty_cells(grid)
    solve_sudoku(grid, empty_cells)

def get_empty_cells(grid):
    empty_cells = []
    for row in range(9):
        for col in range(9):
            if grid[row][col] == 0:
                empty_cells.append((row, col))
    return empty_cells

def solve_sudoku(grid, empty_cells):
    if not empty_cells:
        return True

    row, col = empty_cells[0]
    nums = list(range(1, 10))
    random.shuffle(nums)

    for num in nums:
        if is_valid(grid, row, col, num):
            grid[row][col] = num
            if solve_sudoku(grid, empty_cells[1:]):
                return True
            grid[row][col] = 0

    return False

def is_valid(grid, row, col, num):
    return (
        is_valid_row(grid, row, num) and
        is_valid_column(grid, col, num) and
        is_valid_box(grid, row, col, num)
    )

def is_valid_row(grid, row, num):
    return num not in grid[row]

def is_valid_column(grid, col, num):
    return num not in [grid[row][col] for row in range(9)]

def is_valid_box(grid, row, col, num):
    start_row = (row // 3) * 3
    start_col = (col // 3) * 3
    for i in range(3):
        for j in range(3):
            if grid[start_row + i][start_col + j] == num:
                return False
    return True

def remove_cells(grid):
    num_cells_to_remove = random.randint(40, 50)
    cells = [(row, col) for row in range(9) for col in range(9)]
    random.shuffle(cells)
    for i in range(num_cells_to_remove):
        row, col = cells[i]
        grid[row][col] = 0

def print_grid(grid):
    for row in grid:
        print(row)

# Generate and print a Sudoku grid
grid = generate_grid()
print_grid(grid)
```
